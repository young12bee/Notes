###Chapter 1 - Network Basics and Terminology
- Understanding Network Devices and Cabling
    - Network Devices
        -Hub
            - older networking device
            - connect all the systems together in a network environment
            - layer-1 device
            - receives a signal from 1 system and sends the signal to all other ports on th hub.
            - drawback; uses up bandwidth by sending th data to every port on the hub
            - security issue
            - creates a "shared" network segment that all systems have access to.  Cause domain collision
            
         - Switch
            - connect all systems together in a network environment
            - layer-2 device that filers traffic by the layer-2 address (MAC address or hardware adddress that is assigned to the network card by manufacture
            - The switch is able to filter th traffic because it stores the MAC address of each system connected to the switch and what port that system is connected to in the MAC address table. 
             - MAC address table is a table stored in memory on the switch and is responsible for tracking what ports each system is connected to
             - Collision Domains
                - Switch creates a collision domain that is its own network segment.  Because no other system is on the network segment, there won't be data collision.
                - a group of network systems that share the same network segment and therefore can have their data collide with one another.
              - Benefits:
                1. Filterring
                2. Port mirroring or port monitoring
                    - feature allows the administrator to copy traffice from other ports to a single destination port.
                    - interface port on the switch to monitor traffice sent or received on ports 1 to 5. 
                3. Port security
                    - Controls which system can connect to the switch because the switch can temporariy disable the port
                4. Disabling ports
           - VLANs
               - Virtual LANs
               - create multiple networks within 1 network swithc.
               - placing ports on the switch into groupings
               - When a system is connected to a port on the switch, it becomes a member of the VLAN that the port is associated with.
               - cannot communicate across VLAN without a router
           - Router
               - layer-3 device of the OSI
               - responsible for routing or sending data from one network to another network.
               - use a routing table that resides in memory to determine the networks that the router knows how to send data to
               - network devices that define the boundary of the network by creating what is called a broadcast domain.  A broadcast message is a message that is destined for all systems.
               - is strategically placed on the network to keep those broadcast messages within your network because broadcast traffic does not pass through the router.
           - Load Balancer
               - device that is deisnged to split the load between components such as servers or routers.
               - improve performance
           - Coaxial Cable
               - TV
               - resistant to outsied interference
               - 2 types of coax cable
                   1. Thinnet - for short-distance communication
                       - flexible enough facilitate maneuvering between workstations.
                       - connects directly to a workstation network adapter card by using a British naval connectoer (BNC)
                   2. Thicknet - support data transfer over longer distances.
                       - backbone to connect several smaller thinnet-based networks
             - Twisted-Pair Cable
                 - help reduce crosstalk or interference from outside electrical devices.
                 1. Unshielded Twisted-Pair (UTP) Cable
                     - distance 100 meters
                 2. Straight-Through Cables  CAT5 UTP
              - Crossover Cable
              - Shielded Twisted-Pair (STP) Cable
              - Fiber-Optic Cable
                  - carries digital data signals
                  - 2 types of fiber-optic 
                  1. Single-mode fiber
                  2. Multimode fiber
              - supports up to 1000 stations and can carry the signal up to and beyond 2 kilometers.  
              - highly secure
                  
- Understanding TCIP/IP
    - IP Address
        - 32-bit value; identifies the system on the network (Internet)
        - 4 decimal values separated by decimal points
        - each decimal values (octet); 4 octets in IP address
        - divided in to 2 parts:
            1. network ID - assigns a unique address to the network itself
            2. host ID - identifies the system on the network
     - subnet mask determines the number of bits that make up the network ID and the number of bits that make up the host ID.
