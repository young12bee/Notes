### TWO-MINUTE DRILL
- Understanding Network Devices and Cabling
    - Switches should be used instead of hubs
        1. filter traffic by sending data only to the port on the switch that destination system resides on
        2. security feature (advantages)
            - disable unused ports
            - configure port security 
                - control what systems can connect to port by MAC address
     - Hubs 
         1. sends data to all ports on the hub
         2. allows someone to view all network trafic
         
     - Routers
         1. layer-3 devices
         2. route data from one network to another
         3. to break a network into multiple broadcast domains
      
     - Proxy server 
         1. send outbound Internet requests on behalf of clients 
         2. used to filter the web sites a user can visit and the types of applications thy can use for outbound communication.
         
     - Coax and twisted-pair cabling
         1. copper core to carry an electrical signal
         2. fiber-optic cabling used to carry pulses of light
         3. fiber-optic (highly secre environments)
         
- Understanding TCP/IP
    - IP protocol - responsible for logical addressing and routing
    - TCP protocol 
        - reliable delivery
        - 3-way handshake
        - guarantees delivery by using sequence numbers
        - acknowledgement numbers
        - use port number to identify sending and receiving application of the data
        - uses flags to identify important types of packets
            - SYN
            - ACK
            - FIN
            - RST
            - URG
            - PSH
    - UDP (User Datagram Protocol)
        - connectionless communication
        - streaming
        - NOTE: Security+ (Memorize common ports)  
        
    - ICMP (Internet Control Message Protocol)
        - status and error reporting
        - uses types and codes to identify different types of messages
        - Type 8 is used for echo request
        - Type 0 is used for echo reply
        
    - ARP (Address Resolution Protocol)
        - converting logical address (IP address) to physical address (MAC address).
        
- Network Security Best Practices
    - fiber-optic cabling -
        - difficult to tap
        - immune to electrical interference
    - secure versions of protocols to encrypt communication
        - SSH, SCP, HTTPS, FTPS
    - use port security switch
        - control which systems can connect to the switch
    - Disable all unused ports on the switch

### CERTIFICATION OBJECTIVE 1.02
- UNDERSTANDING TCP/IP
    - TCP/IP
        1. IP address
        2. subnet mask
        3. default gateway
        
    - IP Address
        - 32-bit value identifies the system on the network (Internet)
        - consists of 4 decimal values separated by decimal
        - each values is 8 bits
        - 8 bits x 4 = 32-bit address
        - each decimal values (octec)
        - 4 octets = IP address
        - 2 parts
            1. network ID
            2. host ID
        - subnet mask determines
            - the number of bits makes up the network ID
            - the nubmer of bits makes up the host ID
            
     - Subnet Mask
         - determine if target computer exists on the same network
         - same network IDs, no router needed to transfer data
         - different network IDs, data will need to be passed to the router so the router can send the data to other network.
         
     - How data are transfer
         - The system compares the destination system's IP address to its own network address.  If they have the same network ID, the data is sent directly from your system to the destination system.  If the two systems are on different networks, your system must pass the data to the router so that router can send the data to the destination system's router.  Your system uses the default gateway IP address router to send data from your network.
     
     - Default Gateway
         - The default gateway is the IP address of the router that can send data from your network.
